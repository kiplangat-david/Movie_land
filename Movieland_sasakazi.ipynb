# **Data Description**

Our data for this project was taken from Kaggle. This is the [link](https://www.kaggle.com/datasets/harshitshankhdhar/imdb-dataset-of-top-1000-movies-and-tv-shows?resource=download) to the dataset


# **Data Features**

-Our dataset has 13 features. Below are the descriptions of the features

1. **Poster_Link** - Link of the poster that imdb using

2. **Series_Title** - Name of the movie

3. **Released_Year** - Year at which that movie released

4. **Certificate** - Certificate earned by that movie

5. **Runtime** - Total runtime of the movie

6. **Genre** - Genre of the movie

7. **IMDB_Rating** - Rating of the movie at IMDB site

8. **Overview** - mini story/ summary

9. **Meta_score** - Score earned by the movie

10. **Director** - Name of the Director

11. **Star1,Star2,Star3,Star4** - Name of the Stars

12. **No_of_votes** - Total number of votes

13. **Gross** - Money earned by that movie

#Question 1.
# Which movie had the highest rating?

```python
#Q1.
#Which movie had the highest rating?
top_rating = imdb[['Series_Title','IMDB_Rating']].sort_values(by = ['IMDB_Rating'], ascending=False)
top_rating.head(1)
```

#Question 2
#How many reviewers gave it the highest rating?

```python
#Q2
#How many reviewers gave it the highest rating?


#we will locate the movie with the highest rating and check the No_of_ Votes.
#The numbeer of vots here represent the viewers who rated the movie
reviewers = imdb[['Series_Title','IMDB_Rating','No_of_Votes']].sort_values(by = ['IMDB_Rating'], ascending=False)
reviewers.head(1)


#We can also locate the rating and the number of votes
reviewers.iloc[:,[1,2]].head(1)
```
